# 第6章: 英語テキストの処理
英語のテキスト[nlp.txt](http://www.cl.ecei.tohoku.ac.jp/nlp100/data/nlp.txt) に対して、以下の処理を実行せよ。

### 50.文区切り
(. or ; or : or ? or !) → 空白文字 → 英大文字というパターンを文の区切りと見なし，入力された文書を1行1文の形式で出力せよ．  
メモ: ()で括ることによりグループ化を行う。また\numberでマッチしたものにアクセスできる

```python
import pydot_ng as pydot
import re
from nltk.stem.porter import PorterStemmer
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
nlp_text = open('./resource/nlp.txt','r')
nlp_sentence_split = open('./resource/nlp_sentence_split.txt', 'w')
for i,line in enumerate(nlp_text):
    if line == '\n':
        continue
    temp_str = re.sub(r"([\.;:!\?])\s+([A-Z])", r"\1\n\2", line)
    temp_str = temp_str.strip()
    itel_str = temp_str.split('\n')
    for j, row in enumerate(itel_str):
        print("{}-{}: {}".format(i,j,row.strip()))
        nlp_sentence_split.write(row+"\n")
nlp_text.close()
nlp_sentence_split.close()
```

### 51.単語の切り出し　
空白を単語の区切りとみなし、50の出力を入力として受け取り、1行1単語の形式で出力せよ。ただし、文の終端では空行を出力せよ

```python
nlp_sentence_split = open('./resource/nlp_sentence_split.txt', 'r')
nlp_word_split = open('./resource/npl_word_split.txt', 'w')
for line in nlp_sentence_split:
    WordList = line.split()
    for word in WordList:
        print(re.sub(r"[\(\),.:;\?!\"\'\-]", '',word))
        nlp_word_split.write(re.sub(r"[\(\),.:;\?!\"\'\-]", '',word)+"\n")
    nlp_word_split.write("\n")
    print('\n')
nlp_sentence_split.close()
nlp_word_split.close()
```

### 52.  ステミング
51の出力を入力として受け取り、Porterのステミングアルゴリズムを適用し、単語と語幹をタブ区切り形式で出力せよ。Pythonでは、Porter のステミングアルゴリズムの実装として[stemming](https://pypi.python.org/pypi/stemming)モジュールを利用すると良い

```python
nlp_word_split = open('./resource/npl_word_split.txt', 'r')
Stemmer = PorterStemmer()
for word in nlp_word_split:
    if word == '\n':
        print("\n\t---Sentence END---\n")
        continue
    temp_word = word.strip()
#     print(word)
    print("Word: {}\tStem: {}".format(temp_word, Stemmer.stem(temp_word)))
```

### 56. 共参照解析
Stanford Core NLPの共参照解析の結果に基づき，文中の参照表現（mention）を代表参照表現（representative mention）に置換せよ．ただし，置換するときは，「代表参照表現（参照表現）」のように，元の参照表現が分かるように配慮せよ．

```python
nlp_xml = open('./resource/nlp.txt.xml', 'r')
soup = BeautifulSoup(nlp_xml.read(), "xml")
iter_soup = soup.find("coreference")
Replace_List = []
for i in iter_soup.find_all("coreference"):
    RepresentativeMention = i.find_all("mention")[0].find("text").get_text()
#     RepresentativeMention = re.sub('-LRB-','(',RepresentativeMention)
#     RepresentativeMention = re.sub('-RRB-',')',RepresentativeMention)
    for mention in i.findAll("mention")[1:]:
        # [Sentence, Start, End,Text]
        Sentence = int(mention.find("sentence").get_text())
        Start = int(mention.find("start").get_text())
        End = int(mention.find("end").get_text())
        Text = mention.find("text").get_text()
        TempList = [Sentence, Start, End, RepresentativeMention, "【"+Text+"】"]
        Replace_List.append(TempList)

Sentences = []
for sentence in soup.find("sentences").find_all("sentence"):
    tempSentence = sentence.find_all("word")
    func = lambda x:x.get_text()
    tempSentence = list(map(func, tempSentence))
    Sentences.append(" ".join(tempSentence))

Words = list(map(lambda x:x.split(), Sentences))    

for l in Replace_List:
    sentenceID = l[0]
    startTokenID = l[1]
    endTokenID = l[2]
    replaceString = l[3]+l[4]
    
    tempWords = Words[sentenceID-1]
    tempWords[startTokenID] = replaceString
    
    for i in range(startTokenID+1, endTokenID):
        tempWords[i] = ''
    Words[sentenceID-1] = tempWords

    
Words2Sentence = lambda x:" ".join(x)
for l in list(map(Words2Sentence, Words)):
    temp = re.sub(r'\s+', ' ', l)
    temp = re.sub('-LRB-','(',temp)
    temp = re.sub('-RRB-',')',temp)
    print(temp) 
```

### 57. 係り受け解析
Stanford Core NLPの係り受け解析の結果（collapsed-dependencies）を有向グラフとして可視化せよ．可視化には，係り受け木をDOT言語に変換し，Graphvizを用いるとよい．また，Pythonから有向グラフを直接的に可視化するには，pydotを使うとよい．

```python
nlp_xml = open('./resource/nlp.txt.xml', 'r')
soup = BeautifulSoup(nlp_xml.read(), "xml")
sentencesSoupList = soup.find("sentences").find_all("sentence")

for i,sentenceSoup in tqdm(enumerate(sentencesSoupList)):
# sentenceSoup = sentencesSoupList[0]
    graph = pydot.Dot(graph_type = "digraph")
    Nodes = [pydot.Node("0.【ROOT】")]
    Edges = []
    title = "resource/graphs/sentence{}.png".format(i+1)
    for j,word in enumerate(sentenceSoup.findAll("word")):
        tempNode = pydot.Node(str(j+1)+".【"+word.get_text()+"】")
        Nodes.append(tempNode)
        graph.add_node(tempNode)

    hoge = sentenceSoup.find_all("dependencies")
    for i in hoge:
        if i.attrs['type']=="collapsed-dependencies":
            for depSoup in i.find_all("dep"):
                src = int(depSoup.find("governor").attrs['idx'])
                dst = int(depSoup.find("dependent").attrs['idx'])
                tempEdge = pydot.Edge(Nodes[src],Nodes[dst])
                Edges.append(tempEdge)
                graph.add_edge(tempEdge)
    graph.write_png(title)
```

### 58. タプルの抽出
Stanford Core NLPの係り受け解析の結果（collapsed-dependencies）に基づき，「主語 述語 目的語」の組をタブ区切り形式で出力せよ．ただし，主語，述語，目的語の定義は以下を参考にせよ．

* 述語: nsubj関係とdobj関係の子（dependant）を持つ単語
* 主語: 述語からnsubj関係にある子（dependent）
* 目的語: 述語からdobj関係にある子（dependent）

```python
nlp_xml = open('./resource/nlp.txt.xml', 'r')
soup = BeautifulSoup(nlp_xml.read(), "xml")
sentencesSoupList = soup.findAll("dependencies")
collapsedSoupList = []
for l in sentencesSoupList:
    if l.attrs['type'] == "collapsed-dependencies":
        collapsedSoupList.append(l)
for i, temp_collapsedSoup in enumerate(collapsedSoupList):
    nsubjs = []
    dobjs = []
    for depSoup in temp_collapsedSoup.findAll("dep"):
        if depSoup.attrs['type'] == 'nsubj':
            nsubjs.append([depSoup.find("governor").get_text(),depSoup.find("dependent").get_text()])
        if depSoup.attrs['type'] == 'dobj':
            dobjs.append([depSoup.find("governor").get_text(),depSoup.find("dependent").get_text()])

    for nsubj in nsubjs:
        for dobj in dobjs:
            if nsubj[0] == dobj[0]:
                print("Sentence{}:\t{}\t{}\t{}".format(i, nsubj[1], nsubj[0], dobj[1]))
```

### 59. S式の解析
Stanford Core NLPの句構造解析の結果（S式）を読み込み，文中のすべての名詞句（NP）を表示せよ．入れ子になっている名詞句もすべて表示すること．

```python
nlp_xml = open('./resource/nlp.txt.xml', 'r')
soup = BeautifulSoup(nlp_xml.read(), "xml")
S_expressionList = soup.findAll("parse")
```

```python
# NP_pattern = re.compile(r'(\(NP\s(\(JJ|NN|NNS)\s[A-Za-z]*\))+\)')
NP_pattern = re.compile(r"(\(NP\s(\((JJ|NN|NNS)\s[A-Za-z]+\)\s*)(\((JJ|NN|NNS)\s[A-Za-z]+\)\s*)+\))")
```

```python
ret_List = []
for S_expression in S_expressionList:
    NP_List = NP_pattern.findall(S_expression.get_text())
    NP_List = list(map(lambda x:re.sub('NNS|NP|JJ|NN|\(|\)','',x[0]).split(), NP_List))
    ret_List.append(NP_List)
```
