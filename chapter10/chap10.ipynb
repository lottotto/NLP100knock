````python
def getVector(token):
    temp_index = dict_t_keys.index(token)
    return X300dim[temp_index]

def calc_cossim(vec1, vec2):
    inner_product = np.dot(vec1,vec2)
    norm1 = np.linalg.norm(vec1)
    norm2 = np.linalg.norm(vec2)
    return inner_product/(norm1*norm2)

def SimilarWords(vec):
    result_List = []
    for i, key in enumerate(dict_t_keys):
        if len(result_List) < 11:
            result_List.append([key, calc_cossim(vec, X300dim[i])])
            result_List = sorted(result_List,key=lambda x:x[1],reverse=True)
        else:
            temp_cossim = calc_cossim(vec, X300dim[i])
            if result_List[-1][1] < temp_cossim:
                result_List[-1] = [key, calc_cossim(vec, X300dim[i])]
                result_List = sorted(result_List,key=lambda x:x[1], reverse=True)
    return ret_list[1:]
````

### 96. 国名に関するベクトルの抽出
word2vecの学習結果から国名に関するベクトルのみを抜き出せ

````python
ret_list = []
with open('./resource/country.txt','r') as country_file:
    for line in country_file:
        if len(line.split())>1:
            temp = "_".join(line.split())
            ret_list.append(temp)
        else:
            ret_list.append(line.rstrip())
ret_list
````

````python
def country2vec(x):
    try:
        return model.wv.__getitem__(x)
    except:
        return None
    
vec_list = list(map(country2vec, ret_list))
country_vec = list(zip(ret_list,vec_list))
````

````python
temp = country_vec.copy()
for country, vec in country_vec:
    if vec is None:
        temp.remove((country,vec))
        
country_vec = temp
````

### 98. Ward法によるクラスタリング
96の単語ベクトルに対して，Ward法による階層型クラスタリングを実行せよ．さらに，クラスタリング結果をデンドログラムとして可視化せよ．

````python
from scipy.cluster.hierarchy import linkage, dendrogram
Z = linkage(countryDF.T, 'ward')
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(Z,leaf_rotation=90.,
leaf_font_size=8.)
plt.show()
````
